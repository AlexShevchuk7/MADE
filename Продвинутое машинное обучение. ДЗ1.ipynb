### Часть 1. О Теореме Байеса. 

1. Ошибка прокурора. 

Введем обозначния:

Собыите $G$ - человек виновен

Событие $\neg G$ - не виновен

Событие $b_i$ - человек имеет *i-ую* группу крови

Прокурор оценил вероятность 

Тогда, полная вероятность быть виновным выражается, как

$$P(G) = \sum_{i}P(G,\,b_i) = \sum_{i}P(G\,|\,b_i)\cdot P(b_i)$$

По условию задачи, у убийцы была найдена редкая группа крови. Обозначим ее за $b_r$. Тогда, условные вероятности $P_{i \neq r}(G\,|\,b_i) = 0$ и вероятность быть виновным при условии найденных доказательств оказывается равной $P(G\,|\,E_r) = P(G\,|\,b_r)\cdot P(b_r), \, где \, E_r \,$ - улика, состаящая в том, что у убийцы редкая группа крови $r$.

Тот же вывод может быть сделан из теоремы Байеса. Найдем вероятность оказаться виновным при условии обладания *r-ой* редкой группой крови $P(G\,|\,b_к)$:

$$P(G\,|\,b_r) = \frac{P(b_r\,|\,G)\cdot\,P(G)}{P(b_r)},$$

где $P(b_r\,|\,G)$ - вероятность иметь группу крови $r$ при условии, что ты виновен. Так как следствие доказало, что убийцы обладал именно этой группой крови, то по условию задачи эта вероятность равна 1, т.е. если ты виновен, то твоя группа крови с вроестностью 1 будет именно *r-ой*.

Тогда вероятность оказаться виновным для обладателя той же группы крови составит 

$$P(G\,|\,b_r) = \frac{P(G)}{P(b_r)},$$ где $P(G)$ - априорная вероятность быть виновным в совершении подобного преступления, т.е. вероятность, что некий наугад взятый человек мог совершить подобное преступление. 

Так как данной группой крови обладает всего 1% населения, а население города составляет $10^{6}$, то вероятность оказаться виновным для произвольно взятого человека составляет $10^{-6}$ и 

$$P(G\,|\,b_r) = \frac{P(G)}{10^{-2}} = \frac{10^{-6}}{10^{-2}} = 10^{-4}$$

Таким образом, данная ваероятность является вероятностью оказаться виновным для произвольно взятого человека с $r-ой$ группой крови.

Прокурор по условию задачи оценивал "Шанс, что у подсудимого была бы именно такая группа крови, если бы он был невиновен".

В обозначениях теории вероятностей это можно выразить, как $P(b_r\,|\,\neg G)$, а должен был оценивать вероятность того, что 
подсудимый виновен при условии, что у него такая же группа крови, т.е. $P(G\,|\,b_r)$.

Вероятность $P(b_r\,|\,\neg G)$ можно оценить исходя из того, что 

$$P(b_r) = P(b_r\,|\,G)\cdot P(G) + P(b_r\,|\,\neg G)\cdot P(\neg G)$$

Так как $P(G) + P(\neg G) = 1$, а $P(b_r\,|\,G) = 1$ по условию задачи, то

$$P(b_r) = 10^{-2} = P(G) + P(b_r\,|\,\neg G)\cdot (1 - P(G))$$

Тогда вероятность, которую оценивал прокурор можно выразить как 

$$P(b_r\,|\,\neg G) = \frac {10^{-2} - P(G)}{1 - P(G)} = \frac {10^{-2} - 10^{-6}}{1 - 10^{-6}} \approx 0.01$$ 

2. Ошибка адвоката

Адвокат оценил вероятность $P(G\,|\,b_r, E_r)$, которая соответствует вероятности произвольно выбранному человеку с $r-ой$ группой крови оказаться виновным, если преступник также проживает в этом городе. Если не существует других обстоятельств, указывающих на преступника, то адвокат правильно оценил вероятность оказаться виновным при условии обнаруженного свидетельства. Однако, если существует дополнительное свидетельство, то указанная вероятность является всего лишь априорной.


3. По формулме Байеса вероятность оказаться больным при условии положительного сданного теста равна: 
    
$$P(D\,|\,T_+) = \frac{P(T_+\,|D)\cdot P(D)}{P(T_+\,|\,D)\cdot P(D) + P(T_+\,|\,\neg D)\cdot P(\neg D)},$$

где

$P(D)$ - априорная вероятность иметь заболевание, равная 0.01 по условию задачи

$P(T_+\,|D)$ - условная веротяность сдать положительный тест при условии, что ты болен, по условию задачи равная 0.95

$P(T_+\,|\,\neg D)$ - условная вероятность сдать положительный тест, при условии, что ты не болен, равная в задаче 0.05

$P(\neg D)$ - априорная вероятность не иметь заболевания, в задаче - 0.99

Тогда вероятность того, что ты болен при условии, что тест показал положительный результат:  

$$P(D\,|\,T_+) = \frac{0.95 \cdot 0.01}{0.95 \cdot 0.01 + 0.05 \cdot 0.99 } = \frac{0.0095}{0.059} \approx 0.161$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style("whitegrid")
sns.set_palette("colorblind")
palette = sns.color_palette()

from matplotlib import rc
rc('font',**{'family':'sans-serif'})
rc('figure', **{'dpi': 100})
```

По условию задачи построим две функции. Первая будет считать вероятность $P(D\,|\,T_+)$ в зависимости от вероятности выдать положительный тест действительно больному (рост этой вероятности отвечает за уменьшение числа ложноотрицательных тестов): 

$$P(D\,|\,T_+) = \frac{(0.95 + \alpha) \cdot 0.01}{(0.95 + \alpha) \cdot 0.01 + 0.05 \cdot 0.99 },$$

где $\alpha$ изменяется в пределах от 0 до 0.05, что соответствует отсутсвию изменений в точности тестов ($\alpha = 0$) и
увеличение точности до 100% ($\alpha = 0.05$).

Вторая функция будет считать вероятность $P(D\,|\,T_+)$ в зависимости от вероятности выдать положительный тест здоровому (снижение этой вероятности отвечает за уменьшение числа ложноположительных тестов): 

$$P(D\,|\,T_+) = \frac{0.95 \cdot 0.01}{0.95 \cdot 0.01 + (0.05 - \alpha) \cdot 0.99 },$$

где $\alpha$ также изменяется в пределах от 0 до 0.05, что соответствует отсутсвию изменений в точности тестов ($\alpha = 0$) и сведение числа ложноположительных тестов к 0 ($\alpha = 0.05$).

Таким образом, $\alpha$ для обеих функций имеет одинаковый масштаб, что позволяет удобно оценить полезность внесения изменений в уменьшение числа ложноположительных тестов и в уменьшение числа ложноотрицательных тестов.

```python
def fp_rate(alpha):
    return (0.95 * 0.01) / (0.95 * 0.01 + (0.05 - alpha) * 0.99)

def fn_rate(alpha):
    return ((0.95 + alpha) * 0.01) / ((0.95 + alpha) * 0.01 + 0.05 * 0.99)
```

```python
xm = np.linspace(0, 0.05, 50)
fig = plt.figure(figsize=(10,8))
plt.plot(xm, fp_rate(xm), label="Точность теста при уменьшении ложноположителных")
plt.legend(loc="upper left", fontsize=10)
plt.plot(xm, fn_rate(xm), label="Точность теста при уменьшении ложноотрицательных", color='red')
plt.legend(loc="upper left", fontsize=10)
plt.xlabel(r'$\alpha$', loc='right', fontsize=18)
plt.ylabel(r'$P(D\,|\,T_+)$', loc='top', fontsize=12)
plt.show()
```

Вывод: снижение числа ложноположительных тестов приводит к существенному увеличению точности теста в том смысле, 
в котором он определен выше. В то же время снижение числа ложноотрицательных приводит к едва заметному росту точности теста. Связано это, преждем всего, с редкой встречаемостью заболевания.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from scipy.stats import multivariate_normal

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style("whitegrid")
sns.set_palette("colorblind")
palette = sns.color_palette()
```

```python
# Считаем 50 точек начиная с 03.03.2020

df = pd.read_csv('owid-covid-data.csv')
ru_df = df[df.location == 'Russia']
start_index = ru_df[ru_df.date == '2020-03-03'].index[0]
data = pd.DataFrame(ru_df['total_cases'].loc[start_index : start_index + 49])
data.reset_index(inplace=True, drop=True)
```

```python
# Строим график зависимости

plt.scatter(data.index, data)
```

```python
# Строим график зависимости в логарифмической шкале

plt.scatter(data.index, np.log(data))
```

```python
# Оцениваем коэффициенты w0 и ц1 из линейной регриссии на логарифм от данных:

log_reg = LinearRegression(fit_intercept=False)
x = np.c_[np.ones(50), np.arange(50)]
y = np.log(data.values)
log_reg.fit(x, y)
log_reg.coef_
```

```python
# Строим график зависимости в логарифмической шкале и прямую с полученными коэффициентами

w0_prior, w1_prior = log_reg.coef_[0]
x = np.arange(50)
y = lambda x: w0_prior + w1_prior * x

plt.figure(figsize=(8, 6))
plt.plot(x, y(x), color='purple', label='регрессия на логарифм данных')
plt.scatter(data.index, np.log(data), color='green', label='логарифм данных')
plt.legend(loc='upper left', fontsize=14)
plt.show()
```

```python
# Построим экспоненциальные зависимости: 

plt.figure(figsize=(8, 6))
plt.plot(np.exp(y(x)), color='green', label='кривая по априорной оценке')
plt.scatter(x, data, label='кривая данных')
plt.legend(loc='upper left', fontsize=14)
plt.show()
```

Видно, что кривая, построенная на основе регрессии в логарифмическом масштабе плохо приближает экспоненту

```python
# Построим сетку для оценки коэффициентов w0 и w1

w0_range = np.linspace(w0_prior - 2, w0_prior + 2, 50)
w1_range = np.linspace(0, w1_prior + 2 * w1_prior, 50)
W0, W1 = np.meshgrid(w0_range, w1_range)
```

```python
# Сделаем выбор наилучших коэффициентов, соответствующих минимуму среднеквадратичной ошибки

x = np.arange(50)
y = data['total_cases'].values
N = y.shape[0]
x_range = np.c_[np.ones(50), np.arange(50)]
var = []
err = []

for w0 in w0_range:
    for w1 in w1_range:
        variance = sum((y - np.exp(x_range.dot([w0, w1]))) ** 2) / N
        var.append((w0, w1, variance))
        mse = np.sqrt(variance)
        err.append((mse, w0, w1))
```

```python
w0_prior, w1_prior = min(err, key=lambda x: x[0])[1:]
print(f'Кривая с априорными коэффициентами задается уравнением y = exp({np.round(w0_prior, 3)} + {np.round(w1_prior, 3)} * x)')
```

```python
# Построим график экспоненты с такими коэффициентами. Приближение получилось лучше

plt.figure(figsize=(8,6))
plt.plot(np.arange(50), 
         np.exp(x_range.dot([w0_prior, w1_prior])),
         color='green',
         label='Эмпирическое приближение')
plt.scatter(x, data, label='кривая данных')
plt.legend(loc='upper left', fontsize=14)
plt.show()
```

```python
# Оцениваем априорную дисперсию:
prior_variance = sum((y - np.exp(x_range.dot([w0_prior, w1_prior]))) ** 2) / N
print(prior_variance)
```

Возьмем в качестве априорных значений коэффициента и дисперсии, получвшиеся выше значения: \
$w_0 = 3.821$ \

$\sigma^2 = 652325$

```python
# Определеим функции для Байесовского вывода (вычисление среднего и матрицы ковариаций), а также среднеквадратичной 
# ошибки

def bayesian_update(mu, sigma, x, y, sigma_noise=0):
    x_matrix = np.exp(mu * x)
    
    sigma_n = np.linalg.inv(np.linalg.inv(sigma)+ (1 / (sigma_noise ** 2)) * np.matmul(np.transpose(x_matrix), x_matrix) )
    mu_n = np.matmul(sigma_n, 
                     np.matmul(np.linalg.inv(sigma), 
                               np.transpose(mu)) + (1 / (sigma_noise ** 2)) * np.matmul(np.transpose(x_matrix), y) )
    return mu_n, sigma_n

def mse(x, y, w0, w1):
    return np.sqrt(np.sum((w0 * np.exp(w1 * x) - y) ** 2 / N))
```

```python
# Определим наиболее подходящий коэффициент w1, перебираю значения в диапазоне вокруг его априорного значения, каждый
# раз вычисляя MAP оценок для коэффициента w0 и его дисперсии:

x = np.arange(50)
sigma_noise = np.sqrt(prior_variance)
cur_sigma = 10 * np.array([[1]])
min_mean_squared_error = float('inf')
max_likelihood = 0

for w1 in np.linspace(0.14, 0.18, 100):
    posterior_w, posterior_sigma = bayesian_update([w1], cur_sigma, x, y, sigma_noise=sigma_noise)
    mean_squared_error = mse(x, y, posterior_w, w1)
    if mean_squared_error < min_mean_squared_error:
        min_mean_squared_error = mean_squared_error
        best_w0 = posterior_w
        best_w1 = w1
        best_sigma = posterior_sigma
        print(posterior_w, w1, posterior_sigma, mean_squared_error)
```

#### 3. Моделирование данных о росте заболеваемости нормальными распределениями.
    
Подход к решению: 
Так как мы собрались приближать зависимость кривыми нормальных распределений, то можно попробовать использовать функцию
плотности нормального распределения (гауссиан), которые далее стоит рассматривать в логарифмической шкале для того, 
чтобы приближать зависимость квадратичной функцией. В такой постановке появляется возможность подбора всех коэффициентов
в процессе Байесовского вывода. После получения апостериорных оценок коэффициентов квадратичной функциий необходимо
перевести их в значения среднего и дисперсии нормального распределения, а также выразить дополнительный множитель, 
появляющийся перед интегралом в формуле нормального распределения. Так как мы будем приближать зависимость гауссианом,
то вместо данных из колонки total cases возьмем данные из колонки new cases представялющие из себя производную от первой.
Так как гауссиан сам является производной от функции распределения, то приближение им производной от исходной зависимости
эквивалентно приближению самой зависимости функцией распределения.

План работы:
- загрузить данные из столбца total cases и взять из них выборку из 50 точек по России начиная с 03.03.2020

- обучить Байесовскую регрессию, получив апостериорные оценки коэффициентов

```python
# Считаем 50 точек начиная с 03.03.2020

df = pd.read_csv('owid-covid-data.csv')
ru_df = df[df.location == 'Russia']
start_index = ru_df[ru_df.date == '2020-03-03'].index[0]
data = pd.DataFrame(ru_df['total_cases'].loc[start_index : start_index + 49])
data.reset_index(inplace=True, drop=True)

df = pd.DataFrame(ru_df['new_cases'].loc[start_index : start_index + 49])
df.reset_index(inplace=True, drop=True)
```

```python
# Построим эту же зависимость в логарифмических координатах и подберем 
# коэффициенты квадратичной функции буквально вручную.
# Их и будем использовать в качестве априорных
x = np.arange(50)
f = lambda x: -0.000625 * x**2 + 0.2 * x + 0.5
log_data = np.log(df['new_cases'].replace(to_replace=0, value=1))
plt.figure(figsize=(8, 6))
plt.plot(log_data)
plt.plot(x, f(x))
```

```python
# Убедимся в том, что полученная функция описывает гауссиан:

i = np.arange(300)
f = lambda x: np.exp(-0.000625 * x**2 + 0.2 * x + 0.5)
plt.figure(figsize=(8, 6))
plt.plot(i, f(i))
```

```python
# Посчитаем априорную оценку шума
prior_variance = np.sqrt(np.sum((log_data.values - f(x))**2)/x.shape[0])
prior_variance
```

```python
# Определеим функции для Байесовского вывода (вычисление среднего и матрицы ковариаций)

def bayesian_update(mu, sigma, x, y, sigma_noise=0):
    
    sigma_n = np.linalg.inv(np.linalg.inv(sigma)+ (1 / (sigma_noise ** 2)) * np.matmul(np.transpose(x), x) )
    mu_n = np.matmul(sigma_n, 
                     np.matmul(np.linalg.inv(sigma), 
                               np.transpose(mu)) + (1 / (sigma_noise ** 2)) * np.matmul(np.transpose(x), y) )
    return mu_n, sigma_n
```

```python
# Зададим параметры априорного распределения и расчитаем апостериорные оценки

x = np.arange(50)
x = np.c_[np.ones(50).reshape(50, 1), x.reshape(50, 1), (x * x).reshape(50, 1)]
mu = np.array([0.5, 0.2, -0.000625])
sigma = 10 * np.diag([1, 1, 1])

map_mu, map_sigma = bayesian_update(mu, sigma, x, log_data.values, sigma_noise=prior_variance)
```

```python
# Перейдем от логарифмической шкалы к обычной: 

i = np.arange(50)
exp_g = lambda x, mu: np.exp(np.array([1, x, x**2]).dot(mu))
plt.figure(figsize=(8, 6))
plt.plot(i, exp_g(i, map_mu))
plt.plot(df)
```

```python
# Следуюящая функиця производит пересчет получившихся коэффицентов и считает функцию нормального распределения: 

def normal_distribution(mu):
    a = mu[2]
    b = mu[1]
    c = mu[0]
    sigma = np.sqrt(1 / (2 * abs(a)))
    mu = b / (2 * abs(a))
    C = c - b * b / (4 * a)
    return lambda x: np.sqrt(2 * np.pi) * sigma * np.exp(C) * norm.cdf(x, loc=mu, scale=sigma)
```

```python
# Зависимость и приближение, построенное при помощи нашей функции совместно выглядят так: 

i = np.arange(300)
plt.figure(figsize=(8, 6))
plt.plot(i, normal_distribution(map_mu)(i))
plt.plot(data, color='orange', lw='4')
```

```python
# Семплирования из апостериорного распределения

def plot_sample_lines(mu, sigma, x_range=50, n=1000, points=None):

    w_samples = np.random.multivariate_normal(mu, sigma, n)
    x = np.arange(x_range)

    for weight in w_samples:
        plt.plot(x, normal_distribution(weight)(x), 'k-', alpha=0.1, color='blue')

    if not points is None:
        plt.scatter(points[0], points[1], marker='*')
        
plt.figure(figsize=(8, 6))
plot_sample_lines(map_mu, map_sigma, x_range=70)
plt.plot(data, color='orange', lw='4', label='Приближаемая зависимость')
plt.legend(loc='upper left')
```

```python
def empirical_procentile(mu, sigma, x_range=150, n=2000):
    w_samples = np.random.multivariate_normal(mu, sigma, n)
    x = np.arange(x_range)
    
    a = w_samples[:, 2]
    b = w_samples[:, 1]
    c = w_samples[:, 0]
    C = c - b * b / (4 * a)
    max_curve = w_samples[np.argmax(C)]
    min_curve = w_samples[np.argmin(C)]

    mean_weight = np.mean(w_samples, axis=0)
    max_level = normal_distribution(mean_weight)(max(x))
    ninety_percentile = 0.9 * max_level
    ten_percentile = 0.1 * max_level
    
    plt.plot(x, normal_distribution(mean_weight)(x), 'k-', color='blue', label='Предиктивная кривая')
    plt.plot(x, [ninety_percentile] * x.shape[0], color='red')
    plt.annotate(f'90-й процентиль: {int(ninety_percentile)} человек', (20, 1.02 * ninety_percentile), color='red')
    plt.plot(x, [ten_percentile] * x.shape[0], color='green')
    plt.annotate(f'10-й процентиль: {int(ten_percentile)} человек', (20, 1.1 * ten_percentile), color='green')
    plt.fill_between(x, 
                     normal_distribution(max_curve)(x), 
                     normal_distribution(min_curve)(x),
                     'k-', 
                     color=palette[1], alpha=.3, label="+- дисперсия шума")
    plt.ylim((0, 10e5))
    
plt.figure(figsize=(16, 8))
empirical_procentile(map_mu, map_sigma)
plt.plot(data, color='orange', lw='4', label='Приближаемая зависимость')
plt.legend(loc='upper left')
plt.show()
```
