```python
import re
import random
import numpy as np
import pandas as pd
from itertools import combinations
from collections import defaultdict
from tqdm import tqdm_notebook
```

#### 3. MCMC

```python
# Строим матрицу переходов на основе текста Войны и Мира: 
transition = np.zeros((len(letters), len(letters)))
alphabet = {letter: number for letter, number in zip(letters, range(len(letters)))}

for k, s in enumerate(rutext[1:]):
    transition[alphabet[rutext[k]], alphabet[s]] += 1
    
transition = transition / transition.sum(axis=1)[:, np.newaxis]
```

```python
# Проверим, что по в строках стоят рапсределения вероятностей
transition.sum(axis=1)
```

```python
# Матрица в виде pandas датафрейма
# Символы в столбцах следуют за символом-индексом строки. Поэтому в стооках - распределение вероятности
probs = pd.DataFrame(index=list(letters), columns=list(letters), data=transition)
probs
```

Основная идея МСМС - алгоритма для дешифровки текстов состоит в построении пермутаций шифра (ключа) и вычислением реалистичности (т.е. вероятности того, что получившийся текст схож по структуре с реальным языком) отображения получившегося в результате шифрования закодированного текста новым ключом нового текста, которая представляет из себя следующую вероятность: $\prod_{i=1}^k P({f(s_i), f(s_{i+1}}))$, где $P$ представляет собой вероятности матрицы переходов, а $f$ является отображением из пространства символов в пространство символов, т.е. шифров или ключей, $i$ пробегает по всем символам текста.

Блуждания, таким образом, происходят в пространстве пермутаций ключа дешифрования:
1. Иницализируем пермутацию частотным соответствием символов, взятым из первого задания
2. Меняем ключ переставлением двух различных случайных букв
3. Вычисляем логарифм функции реалистичности (plausibility).
4. Применяем алгоритм Метрополоиса - Гастингса - принимаем пермутацию, если ее вероятность выше, а также в случае если ниже, но при успешном бросании монетки и отклоняем в противном случае.
5. Производим достаточное число итераций.

```python
# Функция поиска нового шифра путем смены двух случайных разных символов местами
def swap(old_cipher):
    
    letters = [key for key in old_cipher.values()]
    
    while True:
        keys = random.sample(range(len(old_cipher)), 2)
        if keys[0] != keys[1]:
            break
    
    letters[keys[0]], letters[keys[1]] = letters[keys[1]], letters[keys[0]]

    new_cipher = {letter: new_key for letter, new_key in zip(old_cipher.keys(), letters)}
    
    return new_cipher

# Функция расчета правдоподобности данных, исходя из нового шифра и таблицы биграмм
def plaus(text, cipher):
    eps = 10e-10
    plausibility = 0
    for k, s in enumerate(text[:-1]):
        plausibility += np.log(probs.loc[cipher[s], cipher[text[k + 1]]] + eps)
    return plausibility

# Функция декодидирования текста по шифру (ключу)
def decode(text, cipher):
    decoded = ''
    for s in text:
        decoded += cipher[s]
        
    return decoded
```

```python
# В качестве функций для случайного блуждания в пространстве шифров воспользуемся адаптированными функциями из лекций
# Значение вероятности принятия / отклонения меньшего значения правдоподобия было подобрано эмпирически
def metropolis_hastings_log_accept(pl, pl_new):
    if pl_new > pl:
        return True
    else:
        return (np.random.rand() < 0.001) #np.exp(pl_new - pl))

def metropolis_hastings(text, iterations, init={letter: letter for letter in letters}):
    cur_cipher = init
    cur_plaus = plaus(text, cur_cipher)
    max_plaus = -10e5
    
    for i in tqdm_notebook(range(iterations)):
        new_cipher = swap(cur_cipher)
        new_plaus = plaus(text, new_cipher)
        if new_plaus > max_plaus:
            max_plaus = new_plaus
            best_cipher = new_cipher

        if (metropolis_hastings_log_accept(cur_plaus, new_plaus)):
            cur_cipher, cur_plaus = new_cipher, new_plaus

    return cur_cipher, cur_plaus, best_cipher, max_plaus
```

```python
# Обучение модели на фрагменте закодированного текста. В качестве начального приближение возьмем частотное соответствие:
cipher, plausibility, best_cipher, max_plaus = metropolis_hastings(encrypted_text[:1000], 15000, init=chars_mapping)
```

```python
# Окончательный шифр и его правдоподобие
best_cipher, max_plaus
```

```python
# Декодированный текст:
decode(encrypted_text, best_cipher)
```
